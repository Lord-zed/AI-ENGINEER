```python
import pandas as pd
import matplotlib.pyplot as plt

electronic_sales = pd.read_excel("/Users/mac/Desktop/data-analysis-and-wrangling/eletronic_sales.xlsx")
electronic_sales


# electronic_sales
```

```python
# Counting of the pc by make

electronic_sales_products = electronic_sales.groupby("Products")["Products"].count()
electronic_sales_products
```

```python
# 

electronic_sales_products.plot(kind='bar')
plt.xlabel('Products')
plt.ylabel('Number of Products')
plt.title('Electronic Sales Products Count by Brand')
```

```python
# Filter the Apple products

filtered_sales = electronic_sales[electronic_sales['Products']=='Apple']
filtered_sales

filtered_apple = electronic_sales.query("Products == 'Apple'")
filtered_apple
```

```python
# Filtered hp
filtered_hp = electronic_sales[electronic_sales['Products']=='HP']
filtered_hp
```

```python
# Filter dell
filtered_dell = electronic_sales.query("Products == 'Dell'")
filtered_dell
```

```python
# Filter lenovo
filtered_lenovo = electronic_sales.query("Products == 'Lenovo'")
filtered_lenovo
```

```python
# Filter compaq
filtered_compaq = electronic_sales.query("Products == 'Compaq'")
filtered_compaq
```

```python
# Converting rows to title()

electronic_sales['Sales Agent'] = electronic_sales['Sales Agent'].str.title()
electronic_sales['Sales Agent']
electronic_sales
```

```python
# Analysing the agent that sold most pcs
electronic_sales["Sales Agent"].unique()

electronic_sales.groupby("Sales Agent").agg({'Products':'sum','Units':'count','Price':'sum'})

```

```python
electronic_sales.set_index('Products',inplace=True)
```

```python

electronic_sales.index.value_counts().head(42).plot(kind = 'bar',figsize=(10,10))
```

```python
# Top  custommers with the highest sales....
# sales["customer_id"].value_counts().sort_values(ascending = False).head(20).plot(kind="bar")

electronic_sales["Sales Agent"].value_counts().plot(kind='bar')
```

```python
# Branches with the highest sales

electronic_sales['Branch'].value_counts().sort_values(ascending=False).plot(kind = 'pie')
```
